```python
#11. Try to calculate the median user score of all the video games
# replace tbd to NaN, then convert string to float to calculate the median
df['User_Score'].replace('tbd', np.nan, inplace=True)
df['User_Score'] = df['User_Score'].astype(float)

gamemedian = df['User_Score'].median()
print('game median value is:', gamemedian)

#replace Nan with median value
df['User_Score'].replace(np.nan, gamemedian, inplace=True)
df
```
